<h1>Creating list of each row form Dictionaries [name] </h1>

```python
import ast
```

```python
 def convert(obj):
     l=[]
     for i in ast.literal_eval(obj):
         l.append(i['name'])
     return l 
         
```

```python
data['genres'] = data['genres'].apply(convert)
```

```python
data['keywords'] = data['keywords'].apply(convert)
```

```python
data.head(2)
```

<h2>Create overview and keywords into list futhter operation</h2>

```python
data['overview'] = data['overview'].apply(lambda x: x.split())
```

```python
data['tagline'] = data['tagline'].apply (lambda x: x.split())
```

```python
data.head(2)
```

<br>

<h2>Removing space  and creating lower case and concanent it for proper tag </h2>

<h3>Removing space</h3>

```python
data['genres'] = data['genres'].apply(lambda x :[i.replace(" ","") for i in x])
data['overview'] = data['overview'].apply(lambda x :[i.replace(" ","") for i in x])
data['tagline'] = data['tagline'].apply(lambda x :[i.replace(" ","") for i in x])
data['keywors'] = data['keywords'].apply(lambda x :[i.replace(" ","") for i in x])
```

<h3>Concatenate at tag (new columns)</h3>

```python
data['tag'] = data['genres'] + data['tagline'] + data['overview'] + data['keywords']
```

```python
movies = data[['id','title','tag']]
```

```python
movies.head(2)
```

<h3>list into string</h3>

```python
movies['tag'] = movies['tag'].apply(lambda x :" ".join(x))
```

<h3>Lower string</h3>

```python
movies['tag'] = movies['tag'].apply(lambda x : x.lower())
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps= PorterStemmer()
```

```python
def stem(txt):
    y=[]
    for i in txt.split():
       y.append(ps.stem(i))
    return " ".join(y)
```

```python
movies['tag'] = movies['tag'].apply(stem)
```
